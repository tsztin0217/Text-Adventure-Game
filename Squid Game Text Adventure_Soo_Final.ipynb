```python
"""
Squid Game Text Adventure

Tsz Tin Vivian Soo
Jan 22nd, 2022

"""

"""
    DocString:
    
    A) Introduction:
    This game is based on the Netflix show Squid Game (2021) written by 
    Hwang Dong Hyuk. 
    
    The show is about 456 players playing childrens game in a twisted way
    in order to compete and win the ultimate cash prize hoping to turn their
    lives around.
    
    The purpose of this game is to have the player re-experience a mini and
    alternate version of the real squid game. 
    
    https://www.netflix.com/title/81040344
    
    It consists of 5 main stages, and also has defined functions
    for start, win, and lose.
    
    This game requires bravery, a YOLO attitude, and having a
    small amount of knowledge from the show will be helpful.
    
    Stage 0: The meeting
    Stage 1: The Slappening
    Stage 2: Honeycomb candy carving
    Stage 3: Tug of War
    Stage 4: Glass Bridge
    Stage 5: Squid Game
    
    B) Known Bugs and/or Errors:
    None.

    
"""
from sys import exit
import random 
# random will be used later to generate red or blue as the lucky color
# should also create different versions of stages
# this is to increase replay value by adding randomness
# I decided to add options throughout the game if the user wants to test the luck
# because the squid game had 456 players, and often times if you don't have the skill
# then you need luck. Sometimes it's not about making the right choices because you don't kniw
# what the correct choices are until you've made them.
# Throughout the game there will be some obvious choices to proceed/fail as well
# some choices that are not so obvious. The point is to make the game not too
# difficult nor too easy

# must define each stage first
def start(): 
    # this is the home page: the first thing that appears when we run the cell
    print("""
    After binging the entire season of Squid Game overnight, you passed out then
    woke up to find yourself reincarnated as one of the candidates for Squid Game.
    
    You're currently sitting at a bench at the subway station.
    
    Heavily in debt, you only have $5 in your pocket and $15 in your bank account.
    
    But rent is due, student loan is due, and your cup noodles are running low,
    so you must come up with a way to make money fast somehow...
    You closed your eyes briefly and prayed for a miracle.
         """)
    input("<Press enter to continue>\n")
    starting_path()

def starting_path(): # this is the first stage
    print("""Please enter your name""")
    
    name = input(prompt="> \n") # name to be used later in the stage
    
    while True:
        age = input(prompt="> What is your age? (Please enter a number)\n")
        ## using try to control for error, This is in case the user inputs a string
        try:
            age = int(age)
            break #if the age entered is an integer, while loop will break and we can
            #move onto the next part of the starting_path() function.
            
        except:
            print("Sorry, please enter your age in number format only")
            input(prompt=">Press enter to continue\n")
            continue # continue to repeat loop when no integer input
            
        
  
    while True:
        debt = input(prompt=">Pick a lucky number from 1 to 10! \n")
        
        try: #control for error
            debt = float(debt)
            if int(debt) <= 10 and int(debt)>=1:
                break
            
            else:
                print("\nPlease enter a number between 1 to 10 only!\n")
                input(prompt=">Press enter to continue\n")
            
        except ValueError: #instead of regular value error message
            print("\nSorry, please enter it in number format only! \n")
            input(prompt=">Press enter to continue\n")
            
             
        except TypeError:
            print("\nSorry, please enter in number format only\n")
            input(prompt=">Press enter to continue\n")
            
        
        except: 
            print("\nSorry, please enter a number that is less than 10\n")
            input(prompt=">Press enter to continue\n")
           
             
    # dynamic string below that includes the inputs entered earlier    
    print(f"""
    A man in a suit approaches you and asks if you would like to play a short mini game 
    with him that will only take a minute.
    "I know your situation, {name.capitalize()}. Age {age}, currently jobless and
    in ${debt} million worth of debt. If you win, I will give you $1,000 in cash right away. 
    If you lose, you will pay me $1,000 instead or it can be substituted with a slap in the face.
    Would you like to play?"
    """)
    input(prompt=">Press enter to continue\n") 
    
    while True: #creating while loop such that if user put in something other than
                #what was prompted, it will allow user to go back to the choices
                #instead of being thrown an error
        choice = input(prompt=f"""> Please enter 1, 2, or 3.  
        1) "Yes, I'm feeling lucky today! I will play."
        2) "No thanks. I'm not interested."
        3) "I am not {name.capitalize()}! You've mistaken me for my evil twin!"\n""")
        
        if "1" in choice or "sure" in choice.lower() or "play" in choice.lower() or "ok" in choice.lower():
            print("\nAwesome! Get ready to play\n")
            input(prompt="<Press enter to continue>\n")
            slap_game()
            break #break the loop when entered this
               

        elif "2" in choice or "no" in choice.lower() or "later" in choice.lower():
            print("""You fared the man well and went home.""")
            fail() # leave game
            break # break loop
        
        # i wanted to add a little fun in the 3rd choice
        # this third choice presents a separate question that gives
        # the user another chance to continue the game
        # by confirming his/her name
        elif "3" in choice or "mistaken" in choice.lower() or "evil" in choice.lower() or "twin" in choice.lower():
            print(f"\nAre you sure you are not {name.capitalize()}?\n")
            evil_twin = input(prompt=f""">
            1. Yes I'm sure
            2. Just kidding. Of course I am {name.capitalize()}. Let's play!
            """)
            if "yes" in evil_twin.lower() or "1" in evil_twin.lower():
                print("""
                "Sorry. My bad!" 
                The man walks away with a suitcase full of cash that could have
                been yours.
                """)
                fail() # fails the game
            elif "no" in evil_twin.lower() or "2" in evil_twin.lower():
                print("""Haha! You got me there! Let's begin!""")
                slap_game() #continue the game
            else:
                print("Sorry please try again") # when entered something else
                # this will trigger the while loop to start the question again

        else:
            print("Sorry, but I couldn't understand. Please try again.")
            input(prompt="<Press enter to continue>\n")
            continue # user entered something other than the prompted choices
        
        

        
def slap_game(): # second stage
    #first creating a variable that's going to be used in a loop later
    # I'm putting times_slapped before the while loop so that each time 
    # when while loop runs this variable doesn't get reset to 0
    times_slapped = 0 

    # making a while loop so that if guessed the wrong color there's infinite retries
    # until the player either guesses right or decides to leave
    while True:
        print("""The game is simple! You just have to pick either red or blue.
              \n""")

        red_blue = ['red', 'blue']
        random_choice = random.choice(red_blue)
    
    # What I'm trying to do here is let the user choose red or blue
    # If the user's choice matches the system's random choice of red or blue,
    # then the user moves on to next scene

    # If the user's choice doesn't match the system's random choice of red or blue,
    # then the user gets slapped since he/she does not have $1,000 to pay,
    # and is then presented with the choice to play the slapping game again 
    # until he/she wins to move onto the next stage or choose walk away to exit the game
        choose = input(prompt='> Choose a color! Red or blue? \n')
            
        if random_choice == choose.lower() or random_choice in choose.lower():
            print("""
            Yay! You've picked the lucky color and received $1,000!
            But $1,000 is still barely enough to repay total amount of debt you owe.
            The man then asks if you would like to play
            a bigger game to win even more prize money!""")
            win_continue = input(prompt="""Yes or No?\n""")
            #this prompts user to either continue or exit the game
            if "yes" in win_continue.lower() or '1' in win_continue:
                print("""
                    You are now officially enrolled into the Squid Game!
                    """)
                stage_2_playground() # will move to next stage
                break # break the while loop
            elif "no" in win_continue.lower() or '2' in win_continue:
                print("""
                You rejected his offer and walked away, not realizing that
                you have missed out on millions worth of prize money.
                """)
                fail()
            # else for in case user didn't enter any of the above
            # to avoid error
            else:
                print("""Sorry I didn't quite get that""")
                win_continue = input(prompt="""Yes or No?\n""")
                if "yes" in win_continue.lower() or '1' in win_continue:
                    print("""
                    You are now officially enrolled into the Squid Game!
                    """)
                    input(prompt="<Press enter to continue>\n")
                    stage_2_playground()
                    break
                if "no" in win_continue.lower() or '2' in win_continue:
                    print("""
                You rejected his offer and walked away, not realizing that
                you have missed out on millions worth of prize money.
                """)
                    fail()
                    break #just making sure while loop doesn't randomly start again


        # this is the elif statement for in case user's guess didn't match 
        # the system's random guess of either blue or red
        elif random_choice not in choose.lower():
            while True:
                times_slapped += 1 
                #every time user doesn't guess correctly, times slapped +1
                print(f"""
    You answered "{choose}", but the lucky color was {random_choice}.
    You lost this mini game and since you don't have $1,000 on you, you were slapped.
    
    You were slapped for a total of {times_slapped} time(s).

    Would you like to continue to play the game?\n""")
                slap_continue = input(prompt=""">
    1. Yes. One more round! 
    2. No. What a scam! I'm outta here. $!%& \n""")

                # added .lower() after the slap_continue so that if user type things
                # such as YeS, it will turn into yes and recognized as valid input
                
                if "yes" in slap_continue.lower() or "1" in slap_continue.lower():
                    break # this will break out of the inner while loop
                elif "no" in slap_continue.lower() or "2" in slap_continue.lower():
                    print("""You've decided it was not worth it and walked away""")
                    fail()
                else:
                    print("""\nPlease choose 1 or 2\n""")                   
        else:
            print("""
            Sorry, please enter a valid command by choosing either red or blue.
            """)

            
def stage_2_playground():
    print("\nWelcome to the playground! Choose a shape provided below!\n")
    input(prompt='<Press any key to continue>\n')
    
    # printing the 3 shape options 
    print("Press 1 for triangle")
    print("Press 2 for star")
    print("Press 3 for umbrella")
    
    choice = input("> ")
    if choice == '1' or "triangle" in choice:
        print("""
        
        Great choice!
        The task is to carve the shape out without breaking the honeycomb candy.
        Since it only involved 3 straight lines, you were able to carve it out with ease.
        
        You may now move onto next stage!
        
        """)
        input(prompt='<Press any key to continue>\n')
        stage_3_tugwar() # moves us to stage 3
        # triangle is the easiest shape to carve out

    elif choice == '2' or "star" in choice:
        print("""
        You were tasked to carve the outline of the honeycomb candy perfectly.
        
        However, due to the difficult shape of the star, you accidentally broke it
        into multiple triangles.
        """)
        fail() # This moves us into fail
    
    # umbrella is the hardest shape, but doesn't mean the player has completely lost hope
    # in the actual show, the main character barely survived by licking his way out 
    
    elif choice == '3' or "umbrella" in choice: 
        print("""You've chosen umbrella! The task is to carve the shape out without 
          breaking the honeycomb candy. What will you do now?""")
        while True:
            print("Press 1 to carve right away")
            print("Press 2 to lick then carve")
            print("Press 3 to pray then carve")

            choice = input('<Please enter 1, 2, or 3\n')
            if "1" in choice or "carve right away" in choice or "right away" in choice:
                print("\nYou're off to the bridge!\n")
                input("<Press enter to continue>\n")
                fail()

            # in the Squid Game show, the main character decided to lick the honeycomb candy
            elif "2" in choice or "lick" in choice or "then" in choice:
                print("""As you lick the honeycomb the edges that were carved melted""")
                input("<Press enter to continue>\n")
                stage_3_tugwar()

            elif "3" in choice or "pray" in choice:
                print("\nWell I guess the quest is over. Good-bye!")
                fail()
            else:
                print("Invalid command. Please try again.")
                input("<Press enter to continue>\n")

        else:
            print("Invalid command. Please try again.")
            input("<Press enter to continue>\n")
            starting_path()
    
    else:
        print("Invalid entry. Please try again.\n")      
        stage_2_playground() 
        # if user doesn't choose any of the shapes or 1 to 3,
        # this will bring us back to the beginning of playground to try again       
        
        
        
        
    

def stage_3_tugwar():
    print("Welcome to stage 3: Tug of War!!!\n")
    print("""You're invited by 3 teams to play tug of war. They're all missing one
    last player. Which team will you choose?""")
    input('<Press any key to continue>\n')

    # presenting options 
    print("1: Join the All Men Team")
    print("2: Join a random group and pray for the best")
    print("3: Join the Smart Team")

    team = ["win", "lose"] # creating a list with either win or lose
    random_team = random.choice(team) # have system randomly select
    
    choice = input("> ")
    
    if '1' in choice or "strong" in choice:
        print("""
        Although the men were strong, they were bad at coordination.
        The opposing team won with a witty strategy.
        You lost.
        """)
        fail() # move to fail
    # joining the random team will subject to random choice of letting the user
    # either win or lose the tug of war
    elif '2' in choice:
        if random_team == "win":
            print("""
            Hooray! Your team was lucky this time got matched against a weak team.
            You've won and may now move on!
            """)
            input('<Press any key to continue>\n')
            stage_4_glass_bridge() # to move on
            
        elif random_team == "lose":
            print("""
            Unfortunately your prayer was not answered.
            Your team was matched against a much stronger team and you lost!
            """)
            fail() # to fail
    elif '3' in choice or "smart" in choice:
        print("""With a strategy, the smart team won! You may move on!""")
        input('<Press any key to continue>\n')
        stage_4_glass_bridge() # to move on
    
    else:
        print("Invalid entry. Please try again.\n")
        stage_3_tugwar() # Brings us back to beginning to try again
    


##################################### STAGE 4 STARTS
    

def stage_4_glass_bridge():
    # instruction
    print("""

    Welcome to the glass bridge!

    In front of you, there are two rows of glass tiles that are each 3 tiles long
    as illustrated below.

             Goal

             1  2
             1  2
             1  2

             Start <- You are here

    To pass the bridge safely, you must step on the correct glass tile, which is
    made of tempered glass and can hold the weight of a hippo. 

    The incorrect glass tile is made of regular glass and will break upon contact.
    Under the current light, all glass tiles look the same.

    There will be one question per step to guide you toward the correct glass tile.
    To answer them, you must choose either the left or right glass tile to step on
    by pressing 1 or 2. 


          """)
    input("Press <Enter> when you are ready")
    
    question_list = ['What is the number 1 principle of Squid Game?',
                     'Which shape does not belong in the Squid Game?',
                     'Which of the two is the better strategy to win?',
                     'In the Squid Game, how many winners can there be in the end?',
                     'Under which condition can Squid Game be terminated?']
    questions = 3
    
    while questions > 0: # while loop under the condition there are more than 0 questions
        # loop will automatically end after going through 3 random questions
        question = random.choice(question_list) # select a random item 
        print(question)
        print(" ")
        
        if question == question_list[0]:
            print("1) Power")
            print("2) Fairness")
            principle = input("\n")
            
            if '2' in principle or 'fair' in principle.lower():
                print("You landed safely on the tempered glass.")
                questions -=1 # correct answer and move on
            else:
                print("Incorrect!")
                print("The glass beneath you breaks and you fell into darkness.")
                fail() # incorrect answer and fail
        
        elif question == question_list[1]:
            print("1) Triangle")
            print("2) Rectangle")
            
            quest = input("> \n")
            
            if '2' in quest or 'rectangle' in quest.lower():
                print("You landed safely on the tempered glass.")
                questions -=1 # correct and move on
            
            else: 
                print("Incorrect!")
                print("The glass beneath you breaks and you fell into darkness.")
                fail()

        elif question == question_list[2]:
            print("1) Cheat")
            print("2) Luck")
            strategy = input("> \n")
            
            if 'cheat' in strategy or '1' in strategy:
                print("Incorrect!")
                print("The glass beneath you breaks and you fell into darkness.")
                fail() # incorrect answer
            
            elif 'luck' in strategy or '2' in strategy:
                print("Correct!")
                print("You landed safely on the tempered glass.\n")
                questions -=1
            
            else:
                print("Incorrect!")
                print("The glass beneath you breaks and you fell into darkness.")
                fail() # if something else was entered
        
        elif question == question_list[3]:
            print("1) One")
            print("2) Two")
        
            how_many = input("> \n")
            how_many = how_many.lower()
        
            if '1' in how_many or 'one' in how_many:
                print("Correct!\n")
                print("You landed safely on the tempered glass.\n")
                questions -= 1
            
            elif '2' in how_many or 'two' in how_many: 
                print("Incorrect!")
                print("The glass beneath you breaks and you fell into darkness")
                fail()
            
        elif question == question_list[4]:
            print("1. When someone gets injured")
            print("2. When majority votes\n")
            
            terminate = input("> \n")
            
            if '1' in terminate or 'someone' in terminate or 'injure' in terminate:
                print("Incorrect!")
                print("The glass beneath you breaks and you fell into darkness")
                fail()
            
            elif '2' in terminate or 'majority' in terminate or 'vote' in terminate:
                print("Correct!\n")
                print("You landed safely on the tempered glass.")
                questions -= 1
        
            
    print("Congrats on passing the Glass Bridge!!\n")
    input("<Press enter to continue>\n")
    squid_game()   
    
#################################### STAGE 4 ENDS    
    
def squid_game(): # final stage 5
    # have user choose someone they dislike to increase incentive to beat the person
    enemy = input("> Name your archnemesis\n")
    print(f"""
    Welcome to the final game: the Squid Game. By winning this game you will
    become the champion and win $4,560,000,000 in total.
    You are up against the other remaining player, {enemy.capitalize()}.
    In this squid-shaped court, you can play either offense or defense.
    Offense wins by reaching the goal past the defense player.
    The defense wins by pushing the offense player out of the court.
    If under any circumstances one player can no longer play, the other player
    automatically wins.
    
    
    """)
        # i'm making a while loop here so that in case something other than 
        # the provided answers gets entered into off_or_def, the game will 
        # allow user to go back and re-enter instead of throwing an error and 
        # exit the game
    while True: 
        off_or_def = input("""
        Would you like to play offense or defense?
        """)
        if 'offense' in off_or_def.lower() or '1' in off_or_def:
            # prompts user input following offense choice
            off_choice = input(""">
            Please choose an action:
            1. kick
            2. ignore
            3. slap
            """)
            if '1' in off_choice or 'kick' in off_choice:
                print(f"""
                {enemy.capitalize()} was about to defend as you dashed toward
                the goal, but immediately dropped unconscious from your sudden
                kick in the crotch.
                """)
                input(">Press <Enter> to continue/n")
                win() # winning option
                break
            
            
            if '2' in off_choice or 'ignore' in off_choice:
                print(f"""
                As you dashed toward the goal, you ignored the risks until
                you were suddenly slapped in the face by {enemy.capitaliza()}
                and everything went dark...
                """)
                fail() # choice 2 is the only fail option
                break
                
            if '3' in off_choice or 'slap' in off_choice:
                print(f"""
                {enemy.capitalize()} was about to defend as you dashed toward
                the goal, but immediately dropped unconscious from your sudden
                slap in the face.
                """)
                input(">Press <Enter> to continue/n")
                win() # winning option
                break
                
                
                
        elif 'defense' in off_or_def.lower() or '2' in off_or_def:
            # prompts user input following defense choice
            def_choice = input(""" 
            Please choose an action:
            1. kick
            2. ignore
            3. slap
            
            """)
            if '1' in def_choice or 'kick' in def_choice:
                print(f"""
                {enemy.capitalize()} was about to break your defense, but
                immediately dropped unconscious from your sudden kick in the crotch.
                """)
                input(">Press <Enter> to continue")
                win() # winning option
                break

            
            if '2' in def_choice or 'ignore' in def_choice:
                print(f"""
                You weren't sure what was going on, until {enemy.capitalize()}
                suddenly slapped you in the face and everything went dark...
                """)
                fail() # choice 2 is the only fail option
                break
                  
            if '3' in def_choice or 'slap' in def_choice:
                print(f"""
                {enemy.capitalize()} was about to break your defense, but
                immediately dropped unconscious from your sudden slap in the face.
                """)
                input(">Press <Enter> to continue/n")
                win()
                break
        else:
            print("""
            Please enter a valid command!
            """)
###################
  
def win(): # winning outcome
    print("""
    Congratulations!! Out of the 456 players, you are the last one standing!!!
    You now have $4,560,000,000 added to your bank account.
    Would you like to try again?
          Press 1 to play again
          Press 2 to end the game
          """)
    
    win_choice = input("> ")
    
    if '1' in win_choice or 'play' in win_choice or 'again' in win_choice:
        start() # goes back to start of the game

    elif '2' in win_choice:
        print("""Thank's for playing!""")
        exit(0)# the game ends
    
    else:
        print("Invalid entry. Please try again.\n")
        win() #return the user to input choices again

    
def fail(): # what happens if at any point the user loses
    print("""\nYou have failed to win the Squid Game. 
          Would you like to try again?
          
          Press 1 to play again
          Press 2 to end the game""")
    
    
    choice = input("> ")
    
    if choice == '1':
        start() # goes back to start of the game

    elif choice == '2':
        print("""
        Thank you so much for playing!!
        """)
        exit(0)# the game ends
    
    else:
        print("Invalid entry. Please try again.\n")
        input('<Press any key to continue>\n')
        
start()


```
